```python
import numpy as np
import gc

# Define folder paths:
DATA_folder  = '../../Data/'
CLUSTER_folder = DATA_folder+'Clusters/'### Data Transformation:
```

```python
# Random Helper Functions:

from math import ceil

def form_seqs(N, max_len):
    seqs = []
    for k in range(ceil(N/max_len)):
        seqs.append((k*max_len, min((k+1)*max_len,N)))
    return seqs;

def form_seqs_v2(lengths):
    seqs = []
    start = 0
    for l in lengths:
        seqs.append((start, start+l))
        start = start+l
    return seqs;
```

### 1) Data Transformation:

```python
### Chunky Transformation Functions (break dataset into chunks - for better memory management)

# Applies a series of different transformations successivly (one on top of eachother)
def apply_transform_successive(data, transforms=[], new_dims=[]):
    # data       - np array of data to be transformed
    # transforms - list of trans. functions to be applied
    # new_dims   - dimension of each transformation (for preallocation)
    
    # Limit the data being transformed to 5000 smpls at time
    max_chunk = 5000 # max number of smpls being fed
    
    n_smpls = data.shape[0]
    
    c_seqs = form_seqs(n_smpls,max_chunk)
    
    new_data = data
    
    if len(transforms)!=0:
        for k in range(len(transforms)):
            current_data = np.empty((n_smpls,new_dims[k]))
            for c_seq in c_seqs:
                current_data[c_seq[0]:c_seq[1],:] = transforms[k](new_data[c_seq[0]:c_seq[1],:])
            new_data = current_data
    
    del current_data
    return new_data;

# Applies a series of different transformations in parallel and appends new data together 
def apply_transform_parallel(data, transforms=[], new_dims=[]):
    # data       - np array of data to be transformed
    # transforms - list of trans. functions to be applied
    # new_dims   - dimension of each transformation (for preallocation)
    
    # Limit the data being transformed to 5000 smpls at time
    max_chunk = 5000 # max number of smpls being fed
    
    n_smpls = data.shape[0]
    
    c_seqs = form_seqs(n_smpls,max_chunk) 
    d_seqs = form_seqs_v2(new_dims)       
    
    if len(transforms)==0:
        new_data = data
    else:
        new_data = np.empty((n_smpls,sum(new_dims)))
        for k in range(len(transforms)):
            d_seq = d_seqs[k]
            trnsf = transforms[k]
            for c_seq in c_seqs:
                new_data[c_seq[0]:c_seq[1],d_seq[0]:d_seq[1]] = trnsf(data[c_seq[0]:c_seq[1],:])
    
    return new_data;

```

```python
from sklearn.pipeline import Pipeline
pipe = Pipeline(steps=[('scal', scaler), ('pca', pca)])
```

```python
### Clustering:

from sklearn.cluster import KMeans
# from numpy.linalg import LinAlgError

def form_clusters(data):
    n_means = 30
    
    kmeans = KMeans(n_clusters=n_means, random_state=0, init='k-means++', algorithm='elkan')
    lbls_kmeans = kmeans.fit_predict(data)
    
    lbls_set = [lbls_kmeans] # just as an option to include more clustering methods
    
    centers   = []
    inv_sq_devs  = []
    
    # Find cluster centers and covar matrix:
    for lbls in lbls_set:
        ig = 0
        for k in range(max(lbls)+1):
            # Ignore single element clusters (cannot determine the spread)
            if(sum(lbls==k)<2):
                #print(k,'ignored (1 el)!!!')
                ig+=1
            else:
                print(k, sum(lbls==k),'elems')
                cluster = data[lbls==k,:]
                centers.append(cluster.mean(axis=0))
                dev = np.std(cluster,axis=0)
                dev[dev==0]+=1e-6
                inv_sq_devs.append(np.reciprocal(np.square(dev)))
        del cluster, dev
    return centers, inv_sq_devs;

# Function to calculate one Fi column for given dataset (on one center)
def calc_fi_column(data, center, inv_sq_devs):
    # data        - given dataset matrix for which to compute fi values
    # center      - given center vector on which to compute fi vals
    # inv_sq_devs - given reciprocal sq. deviation vector on which to compute fi vals 
    
    fi_col = np.empty((data.shape[0],1))
    # tmp = data - np.dot(np.ones((data.shape[0],1)),center.reshape((1,center.shape[0])))
    # tmp2 = np.square(tmp)
    # tmp3 = np.dot(tmp2,inv_sq_devs)
    # fi_col = np.exp(-np.sqrt(tmp3))
    fi_col = np.exp(-np.sqrt(np.dot(np.square(data - np.dot(np.ones((data.shape[0],1)),center.reshape((1,center.shape[0])))),inv_sq_devs)))

    return fi_col;

# Function to compute whole Fi output for given dataset (on all centers)
def fi_transform(data, all_centers, all_inv_sq_devs):
    new_data = np.empty((data.shape[0],len(all_centers)))
    for k in range(len(all_centers)):
        print(k)
        new_data[:,k] = calc_fi_column(data, all_centers[k], all_inv_sq_devs[k])
    
    return new_data
```

```python
all_centers     = []
all_inv_sq_devs = []

# Find centers and covar matrices:
for dig in range(10):
    data = np.load(CLUSTER_folder+'train_'+str(dig)+'.npy')
    new_data = pipe.transform(data)
    print('# Digit ',dig)
    centers, inv_sq_devs = form_clusters(new_data)
    all_centers.extend(centers)
    all_inv_sq_devs.extend(inv_sq_devs)
```

```python
from sklearn.cluster import KMeans

#ns = np.array([1,2,5,10,15,20,25,30,35,40,45,50,60,70,80,90,100,125,150,200,300,400,500,1000])
ns = np.array([1,50,100,200,300,400,500,600,700,800,900,1000])
inertia = np.empty((10,ns.shape[0]))

for dig in range(10):
    data = np.load(CLUSTER_folder+'train_'+str(dig)+'.npy')
    new_data = pipe.transform(data)
    print('# Digit ',dig)
    for k in range(len(ns)):
        kmeans = KMeans(n_clusters=ns[k], random_state=0, init='k-means++', algorithm='elkan')
        kmeans.fit(new_data)
        inertia[dig,k] = kmeans.inertia_
        print(kmeans.inertia_)
        del kmeans
    
```
